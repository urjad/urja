```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver=webdriver.Chrome(r"D:\Internship\chromedriver.exe")
```

```python
# lets first connect to the web driver
url="https://www.amazon.in/"
driver.get(url)
```

```python
search_guitar=driver.find_element_by_id("twotabsearchtextbox")
search_guitar.send_keys("Edible Glitter")
```

```python
search_btn= driver.find_element_by_id("nav-search-submit-button")
search_btn.click()  
```

```python
rating=driver.find_elements_by_xpath("//span[@class='a-size-base']")
rating
```

```python

no_ratings=[]
for i in rating:
    str(i)
    no_ratings.append(i.text)
no_ratings
```

```python
P_name=driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base a-text-normal']")
P_name
Product_name=[]
for i in P_name:
    str(i)
    Product_name.append(i.text)
Product_name
```

```python
Price=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
Price
totalPrice=[]
for i in Price:
    totalPrice.append(i.text)
totalPrice
```

```python
Edate=driver.find_elements_by_xpath("//span[@class='a-text-bold']")
Edate
ExpectedDeliveryDate=[]
for i in Edate:
    ExpectedDeliveryDate.append(i.text)
ExpectedDeliveryDate
```

```python
url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
#@find_element_by_xpath("//p[@class='sc-eYdvao kvdWiq']/a").get_attribute('href')
url
productURL=[]
for i in url:
    productURL.append(i.get_attribute('href'))
productURL
```

```python
print(len(Product_name),len(no_ratings),len(ExpectedDeliveryDate),len(totalPrice),len(productURL))
for i in range(14):
    ExpectedDeliveryDate.append("-")
for j in range(8):
    no_ratings.append("-")
```

```python
amazon=pd.DataFrame({})
amazon['Product Name']=Product_name
amazon['Ratings']=no_ratings[:60]
amazon['Expected Date']=ExpectedDeliveryDate[:60]
amazon['Price']=totalPrice
amazon['Url']=productURL
```

```python
amazon.head()
```

```python
amazon.to_csv('amazon.csv')
driver.close()
```
